## Plotting cell cycle lengths
### Figures 5 and S5

```r
library(ggplot2)
library(ggrastr)
```

```r
data = read.csv("../output/230612_bd396_sphase.csv")
head(data)
```

```r
table(data$tx)
```

```r
# I obtained additional images in June 2023 -- took 4 images per well, originally only processed 1
# but for the samples without sufficient cells I got more 
# A4 is rosc 12.5
# B1 is rosc 25
# C3 is aph 2
```

```r
data = subset(data, !(tx %in% c("ltx500", "ltx1", "ltx2", "tak125", "tak25", 
                                "tak313", "tak5", "tak625", "rosc16", "rosc31")))
data[data$tx == "dmso",]$tx = "DMSO"
data[data$tx == "e2wd",]$tx = "-E2"
data[data$tx == "brq500",]$tx = "BRQ 500 nM"
data[data$tx == "brq1",]$tx = "BRQ 1 µM"
data[data$tx == "brq2",]$tx = "BRQ 2 µM"
data[data$tx == "hu25",]$tx = "HU 25 µM"
data[data$tx == "hu50",]$tx = "HU 50 µM"
data[data$tx == "hu100",]$tx = "HU 100 µM"
data[data$tx == "aph500",]$tx = "APH 500 nM"
data[data$tx == "aph1",]$tx = "APH 1 µM"
data[data$tx == "aph2",]$tx = "APH 2 µM"
data[data$tx == "rosc625",]$tx = "ROSC 6.25 µM"
data[data$tx == "rosc125",]$tx = "ROSC 12.5 µM"
data[data$tx == "rosc25",]$tx = "ROSC 25 µM"

data[data$tx %in% c("a4_4", "a4_2", "a4_3"),]$tx = "ROSC 12.5 µM"
data[data$tx %in% c("b1_4", "b1_2", "b1_3"),]$tx = "ROSC 25 µM"
data[data$tx %in% c("c3_4", "c3_2", "c3_3"),]$tx = "APH 2 µM"
```

```r
data$tx = factor(data$tx, levels=c("DMSO", "-E2", "BRQ 500 nM", "BRQ 1 µM", "BRQ 2 µM",
                                   "HU 25 µM", "HU 50 µM", "HU 100 µM", "APH 500 nM",
                                   "APH 1 µM", "APH 2 µM", "ROSC 6.25 µM", "ROSC 12.5 µM", 
                                   "ROSC 25 µM"))
```

```r
table(data$tx)
```

```r
options(repr.plot.width=6, repr.plot.height=4)
ggplot(data, aes(x=tx, y=slen)) + geom_boxplot(outlier.shape = NA) + 
geom_jitter(color="black", size=0.5, alpha=0.1) + theme_bw() +
scale_x_discrete(guide = guide_axis(angle=45)) + 
ylab("Hours") + xlab("") + ggtitle("S/G2/M phase length")

ggsave("../output/230612_bd396_sg2m_length.pdf", width=5, height=3)
```

```r
data_for_paper = subset(data, tx %in% c("DMSO", "BRQ 1 µM", "HU 50 µM", "ROSC 12.5 µM", "ROSC 25 µM"))
options(repr.plot.width=6, repr.plot.height=4)
ggplot(data_for_paper, aes(x=tx, y=slen)) + geom_boxplot(outlier.shape = NA) + 
rasterize(geom_jitter(color="black", size=0.5, alpha=0.1), dpi=750) + theme_bw() +
scale_x_discrete(guide = guide_axis(angle=45)) + 
ylab("Hours") + xlab("") + ggtitle("S/G2/M length")
ggsave("../output/230615_bd396_sg2m_length_subset.pdf", width=3, height=3)
```

```r
t.test(subset(data, tx == "BRQ 1 µM")$slen, subset(data, tx == "ROSC 25 µM")$slen)
t.test(subset(data, tx == "HU 50 µM")$slen, subset(data, tx == "ROSC 25 µM")$slen)
t.test(subset(data, tx == "HU 50 µM")$slen, subset(data, tx == "ROSC 12.5 µM")$slen)
t.test(subset(data, tx == "HU 50 µM")$slen, subset(data, tx == "ROSC 12.5 µM")$slen)
t.test(subset(data, tx == "DMSO")$slen, subset(data, tx == "ROSC 25 µM")$slen)
```
