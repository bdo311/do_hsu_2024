```python
import gzip
import Levenshtein
import openpyxl
import pandas as pd
import numpy as np
from collections import Counter, defaultdict
from Bio import SeqIO, bgzf
```

```python
allowed_barcodes = pd.read_excel("../data/220922_BD354_barcodes.xlsx")
allowed_barcodes = allowed_barcodes.iloc[:42,:]
allowed_barcodes.head()
```

```python
i1_real = allowed_barcodes["index 1 (P7)"].unique()
i2_real = allowed_barcodes["index 2 (P5)"].unique()
print(i1_real)
print(i2_real)
```

```python
# replace T with G and A with C 
# if I try to replace T with C and G with A, or some other combo, it doesnt work
# so I am convinced this is right
def rep_tg_ac(x):
    x1 = x.replace("T", "G")
    x2 = x1.replace("A", "C")
    return(x2)

i1_gc = [rep_tg_ac(x) for x in i1_real]
i2_gc = [rep_tg_ac(x) for x in i2_real]

print(i1_gc)
print(i2_gc)
```

```python
# testing out algorithm for common barcodes
# barcodes.tsv is from demultiplex algorithm

fn = "../data/barcodes.tsv"

# See if changing T to G and A to C helps
# In case of multiple occurrences of the minimum values, 
# the indices corresponding to the first occurrence are returned.

i1_alreadyseen = set()
i2_alreadyseen = set()
with open(fn, 'r') as ifile:
    for line in ifile:
        z = line.rstrip().split(' ')[1]
        [index1, index2] = z.split('+')
        print(index1, index2)
        
        if index1 not in i1_alreadyseen:
            i1_alreadyseen.add(index1)
            print('i1')
            dist = [Levenshtein.distance(index1, i) for i in i1]
            dist_rep = [Levenshtein.distance(rep_tg_ac(index1), rep_tg_ac(i)) for i in i1]
            
            if np.min(dist) <= 1:
                print(i1[np.argmin(dist)])
            print(dist, dist_rep)
            if np.min(dist) <= 1:
                print(i1[np.argmin(dist)])
            elif np.min(dist_rep) <= 1:
                print(i1[np.argmin(dist_rep)])
            else:
                assert("none found")
        if index2 not in i2_alreadyseen:
            i2_alreadyseen.add(index2)
            print('i2')
            dist = [Levenshtein.distance(index2, i) for i in i2]
            dist_rep = [Levenshtein.distance(rep_tg_ac(index2), rep_tg_ac(i)) for i in i2]
            

            print(dist, dist_rep)
            if np.min(dist) <= 1:
                print(i2[np.argmin(dist)])
            elif np.min(dist_rep) <= 1:
                print(i2[np.argmin(dist_rep)])
            else:
                assert("none found")
```

```python
# require either single mismatch, or perfect match when changing T to G and A to C
def most_likely(bc_obs, bc_set, bc_set_rep, p=True):
    bc_obs_rep = rep_tg_ac(bc_obs)
    dist = [Levenshtein.distance(bc_obs, i) for i in bc_set]
    if p: print(list(zip(bc_set, dist)))
    dist_rep = [Levenshtein.distance(bc_obs_rep, i) for i in bc_set_rep]
    if p: print(list(zip(bc_set_rep, dist_rep)))
    
    if np.min(dist) <= 1:
        return((bc_set[np.argmin(dist)], np.min(dist)))
    elif np.min(dist_rep) == 0:
        return((bc_set[np.argmin(dist_rep)], np.min(dist_rep)))
    else:
        return("none found")  
```

```python
# fn = "/net/bmc-pub17/data/bmc/public/Vanderheiden/220927Van/220930_H7CCT_5952G_L1_NA_sequence_unmapped_barcodes.fastq"
# fn = "/net/bmc-pub17/data/bmc/public/Vanderheiden/220927Van/221005_H7CTY_5960G_L1_NA_sequence_unmapped_barcodes.fastq"
fn = "/net/bmc-pub17/data/bmc/public/Vanderheiden/220927Van/221027_HM5NK_5988G_L1_NA_sequence_unmapped_barcodes.fastq"

i = 0
i1_counts = Counter()
i2_counts = Counter()
with open(fn, 'r') as ifile:
    for line in ifile:
        i += 1
        if i % 10000000 == 0: print(i)
        if i % 4 != 1: continue  # only look at header
        seq = line.rstrip().split(':')[-1]
        [i1, i2] = seq.split('+')
        i1_counts[i1] += 1
        i2_counts[i2] += 1
        
```

```python
# is there an unambiguous best answer for all? 
# yes
```

```python
[print(bc_obs, most_likely(bc_obs[0], i1_real, i1_gc)) for bc_obs in i1_counts.most_common(100)]
```

```python
[print(bc_obs, most_likely(bc_obs[0], i2_real, i2_gc)) for bc_obs in i2_counts.most_common(100)]
```

```python
# get most likely barcodes for all observed barcodes where there is a match
# ignore barcodes that don't have a match

i1_to_bc = {}
for bc_obs in i1_counts:
    z = most_likely(bc_obs, i1_real, i1_gc, p=False)
    if z == "none found": continue
    i1_to_bc[bc_obs] = z[0]
    
i2_to_bc = {}
for bc_obs in i2_counts:
    z = most_likely(bc_obs, i2_real, i2_gc, p=False)
    if z == "none found": continue
    i2_to_bc[bc_obs] = z[0]
```

```python
# open handles
handles = {}

for i in allowed_barcodes.index:
    row = allowed_barcodes.loc[i]
    num = str(int(row["Sample num"]))
    bc1 = row["index 1 (P7)"]
    bc2 = row["index 2 (P5)"]
    
    outgz = bgzf.BgzfWriter(bc1 + "_" + bc2 + "_" + num + ".fastq.gz", "wb")
    handles[bc1 + "_" + bc2] = outgz
    
# writing each batch to file
def write_files(bc_to_fastq, handles):
    for bc in bc_to_fastq:
        SeqIO.write(sequences=bc_to_fastq[bc], handle=handles[bc], format="fastq")
```

```python
# make a dictionary for barcode to fastq records
bc_to_fastq = defaultdict(list)

# save records in dictionary
# fn = "/net/bmc-pub17/data/bmc/public/Vanderheiden/220927Van/220930_H7CCT_5952G_L1_NA_sequence_unmapped_barcodes.fastq"
# fn = "/net/bmc-pub17/data/bmc/public/Vanderheiden/220927Van/221005_H7CTY_5960G_L1_NA_sequence_unmapped_barcodes.fastq"
fn = "/net/bmc-pub17/data/bmc/public/Vanderheiden/220927Van/221027_HM5NK_5988G_L1_NA_sequence_unmapped_barcodes.fastq"

with open(fn, 'r') as handle:
    i = 0
    for record in SeqIO.parse(handle, "fastq-sanger"):
        i += 1
        if i % 1000000 == 0: 
            print(i)
            
            # batch write to file every 1M records 
            write_files(bc_to_fastq, handles)
            bc_to_fastq = defaultdict(list)  # purge
            
        [i1, i2] = record.description.split(':')[-1].split('+')  # read the barcodes
        #print(i1_to_bc[i1], i2_to_bc[i2])
        
        # append records to dict 
        if i1 not in i1_to_bc or i2 not in i2_to_bc: continue  # discard records where both barcodes dont match
        bc_to_fastq[i1_to_bc[i1] + "_" + i2_to_bc[i2]].append(record)  # put in the correct dictionary
        
write_files(bc_to_fastq, handles)
```

```python
# close handles
for i in allowed_barcodes.index:
    row = allowed_barcodes.loc[i]
    num = str(int(row["Sample num"]))
    bc1 = row["index 1 (P7)"]
    bc2 = row["index 2 (P5)"]
    
    handles[bc1 + "_" + bc2].close()
```
