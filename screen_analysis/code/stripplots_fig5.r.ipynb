```r
brq_up_logp = brq_up[,c("GO", "Description", "Log.q.value.", "Enrichment", "X.GeneInGOAndHitList")]
wang_up_logp = wang_up[,c("GO", "Description", "Log.q.value.", "Enrichment", "X.GeneInGOAndHitList")]
combined = merge(brq_up_logp, wang_up_logp, by="GO", all=T)

colnames(combined) = c("GO", "Description1", "BRQ", "BRQ_Enr", "BRQ_Num", 
                       "Description2", "Untreated", "Untreated_Enr", "Untreated_Num")

# each complex has to have at least 7 members in at least one of the top 500 lists to be included
combined = combined[apply(combined[,c("BRQ_Num", "Untreated_Num")],1, function(x) max(x, na.rm=T)) > 5,]
                          
# each point has to be distinct
combined = combined %>% distinct(BRQ, BRQ_Enr, Untreated, Untreated_Enr, .keep_all = TRUE)

# distinct names
#combined = combined %>% distinct(Description1, Description2, .keep_all = TRUE)
                          
# Description1 and Description2 should be merged
combined$Description1 = apply(combined[,c("Description1", "Description2")], 1, function(x) max(x, na.rm=T))
combined$Description2 = combined$Description1
combined$Description3 = combined$Description1

# change NA values
combined$BRQ[is.na(combined$BRQ)] = 0
combined$Untreated[is.na(combined$Untreated)] = 0
                              
combined$BRQ_Enr[is.na(combined$BRQ_Enr)] = 1
combined$Untreated_Enr[is.na(combined$Untreated_Enr)] = 1
                              
# get rid of some labels
combined$Description2[apply(combined[,c("BRQ", "Untreated")], 1, function(x) min(x, na.rm=T)) > -10] = ""
combined$Description2[sapply(combined$Description2, function(x) (nchar(x)) > 35) ] = ""

combined$Description3[apply(combined[,c("BRQ_Enr", "Untreated_Enr")], 1, function(x) max(x, na.rm=T)) < 20] = ""
combined$Description3[combined$Untreated_Enr <= 1] = ""
combined$Description3[sapply(combined$Description3, function(x) (nchar(x)) > 35) ] = ""


dim(combined)
head(combined)
```

```r
# more enriched in BRQ

up = combined[(combined$BRQ_Enr - combined$BRQ) > (combined$Untreated_Enr - combined$Untreated),]
up = up[(up$BRQ < -5) & (up$BRQ_Enr > 5),]
cats = c("ribosome biogenesis", "Cell Cycle", "S Phase", "RNA splicing", "STAGA complex", "Exosome",
         "DNA Methylation", "Activation of ATR in response to replication stress", "Oxidative phosphorylation",
         "Cytosolic iron-sulfur cluster assembly")
up = up[which(up$Description1 %in% cats),]
up = up[order(up$BRQ),]
up[up$Description1 == "Activation of ATR in response to replication stress",]$Description1 = "ATR activation/Replication stress"
up
```

```r
# more enriched in Untreated

down = combined[(combined$BRQ_Enr - combined$BRQ) < (combined$Untreated_Enr - combined$Untreated),]
down = down[(down$Untreated < -5) & (down$Untreated_Enr > 5),]
#down[order(down$Untreated),][50:100,]
cats = c("Mediator complex", "Cellular responses to stress", "Ribosome, cytoplasmic", "Nonsense-Mediated Decay (NMD)",
         "histone modification", "RNA polymerase II core complex", "Nucleotide Excision Repair")
down = down[-4,]  # getting rid of the second mediator entry
down = down[which(down$Description1 %in% cats),]
down = down[order(down$Untreated),]
down
```

```r
row_order = c(up$Description1, down$Description1)

up1 = up[c("Description1", "BRQ", "BRQ_Enr")]
up1$tx = "BRQ"
colnames(up1) = c("Description1", "log_qval", "enr", "tx")
up2 = up[c("Description1", "Untreated", "Untreated_Enr")]
up2$tx = "Untreated"
colnames(up2) = c("Description1", "log_qval", "enr", "tx")

down1 = down[c("Description1", "BRQ", "BRQ_Enr")]
down1$tx = "BRQ"
colnames(down1) = c("Description1", "log_qval", "enr", "tx")
down2 = down[c("Description1", "Untreated", "Untreated_Enr")]
down2$tx = "Untreated"
colnames(down2) = c("Description1", "log_qval", "enr", "tx")

go_df = rbind(up1, up2, down1, down2)
```

```r
options(repr.plot.height=6, repr.plot.width=6)

p1 = ggplot(go_df, aes(x=tx, y=Description1)) + geom_point(aes(size=(-log_qval), color=enr)) + 
scale_y_discrete(limits=rev(row_order)) + theme_minimal() + 
theme(panel.grid.major = element_blank(), text = element_text(size=8*2.35)) +
xlab("") + ylab("") + scale_color_continuous(name="Fold\nenrichment") + 
scale_size("-log10\nq-value", range = c(0, 11)) 

p1

pdf("211228_brq_vs_untreated.pdf", width=7, height=7)
p1
dev.off()
```
