```r
process_res = function(res) {
    tmp <- contrasts.fit(fit, res)
    tmp <- eBayes(tmp)
    res <- topTable(tmp, sort.by = "P", n = Inf)
    return(res)
}
```

```r
conds
```

```r
deps = list()
deps[["brq24"]] = process_res(makeContrasts("condsBRQ24 - condsDMSO24", levels = colnames(coef(fit))))
deps[["brq16"]] = process_res(makeContrasts("condsBRQ16 - condsDMSO24", levels = colnames(coef(fit))))
deps[["brq8"]] = process_res(makeContrasts("condsBRQ8 - condsDMSO24", levels = colnames(coef(fit))))
deps[["e2wd24"]] = process_res(makeContrasts("condsE2wd24 - condsDMSO24", levels = colnames(coef(fit))))
```

## Peak distributions

```r
# get GRanges object for all my peaks
peak_ranges = GRanges(seqnames=peak_info$Chr,
                      ranges=IRanges(start=peak_info$Start,
                                     end=peak_info$End),
                      peak_id=rownames(peak_info))
peak_ranges
```

```r
# annotate all 171k peaks
txdb = TxDb.Mmusculus.UCSC.mm10.knownGene
peakAnno <- as.data.frame(annotatePeak(peak_ranges, tssRegion=c(-2000, 500),
                     TxDb=txdb, annoDb="org.Mm.eg.db"))
peakAnno$annobrief = sapply(peakAnno$annotation, function(x) strsplit(x, ' ') [[1]][1])
peakAnno$annobrief[peakAnno$annobrief == "5'"] = "5' UTR"
peakAnno$annobrief[peakAnno$annobrief == "3'"] = "3' UTR"
peakAnno$annobrief[peakAnno$annobrief == "Distal"] = "Intergenic"
peakAnno$annobrief[peakAnno$annobrief == "Downstream"] = "TTS"
head(peakAnno)
```

```r
# Get baseline frequency across the genome

baseline_freqs = data.frame(table(peakAnno[which(peakAnno$peak_id %in% rownames(brq48)),]$annobrief))
baseline_cts = baseline_freqs
baseline_freqs$Freq = baseline_freqs$Freq/sum(baseline_freqs$Freq)*100
baseline_freqs
```

```r
# Get frequency and counts for 48h

res = brq48
up_peaks_brq = rownames(res[res$logFC > 1.5 & res$adj.P.Val < 0.05,])
down_peaks_brq = rownames(res[res$logFC < -1.5 & res$adj.P.Val < 0.05,])
up_freq = data.frame(table(peakAnno[which(peakAnno$peak_id %in% up_peaks_brq),]$annobrief))$Freq
baseline_freqs$brq_up = up_freq/sum(up_freq)*100
baseline_cts$brq_up = up_freq
down_freq = data.frame(table(peakAnno[which(peakAnno$peak_id %in% down_peaks_brq),]$annobrief))$Freq
baseline_freqs$brq_down = down_freq/sum(down_freq)*100
baseline_cts$brq_down = down_freq

res = e2w48
up_peaks_e2w = rownames(res[res$logFC > 1.5 & res$adj.P.Val < 0.05,])
down_peaks_e2w = rownames(res[res$logFC < -1.5 & res$adj.P.Val < 0.05,])
up_freq = data.frame(table(peakAnno[which(peakAnno$peak_id %in% up_peaks_e2w),]$annobrief))$Freq
baseline_freqs$e2w_up = up_freq/sum(up_freq)*100
baseline_cts$e2w_up = up_freq
down_freq = data.frame(table(peakAnno[which(peakAnno$peak_id %in% down_peaks_e2w),]$annobrief))$Freq
baseline_freqs$e2w_down = down_freq/sum(down_freq)*100
baseline_cts$e2w_down = down_freq

rownames(baseline_freqs) = baseline_freqs[,1]
rownames(baseline_cts) = baseline_cts[,1]
baseline_freqs = baseline_freqs[,2:dim(baseline_freqs)[2]]
baseline_cts = baseline_cts[,2:dim(baseline_cts)[2]]

baseline_freqs
baseline_cts
```

```r
# Fold change observed/expected

freq_change = apply(baseline_freqs[,2:dim(baseline_freqs)[2]],2,function(x) x/baseline_freqs[,1])
colnames(freq_change) = c("BRQ UP", "BRQ DOWN", "-E2 UP", "-E2 DOWN")
freq_change
ggplot(data=melt(log2(freq_change)), aes(x=Var1, y=value)) + geom_col(aes(color=Var2), position="dodge")
                    
```

```r
# log2 genomic localization + plots

freq_change_melted = melt(log2(freq_change[,1:4]))
freq_change_melted$Var1 = factor(freq_change_melted$Var1, 
                                 levels=rev(c("Promoter", "5' UTR", "Exon", 
                                                  "Intron", "3' UTR", "TTS", "Intergenic")))
colnames(freq_change_melted) = c("Feature", "Treatment", "logFC")
head(freq_change_melted)

p = ggplot(data=freq_change_melted, aes(x=Treatment, y=Feature)) + 
    geom_tile(aes(fill=logFC)) + scale_fill_gradient2(
        low = "purple4",
        mid = "white",
        high = "orange4") + 
    theme_bw() + theme(text = element_text(size = 8),
      axis.text=element_text(size=6),
      axis.ticks=element_blank(),
      panel.border = element_blank(),
      panel.grid = element_blank(),
      axis.title=element_blank(),
      legend.key.size = unit(0.2, 'cm'),
      legend.position="bottom",
      legend.title=element_blank(),
      axis.text.x=element_text(angle=45,hjust=1))

p
ggsave("../plots/fig4_genomic_localization.pdf", width=1.5, height=2.3)
```

```r
# get Fisher's p-value for all of these changes

library(stats)
pval = apply(baseline_cts[,2:dim(baseline_cts)[2]],2, function(x) 
    sapply(1:length(x), function(y) fisher.test(
        matrix(c(c(x[y], sum(x)-x[y]),
                 c(baseline_cts[y,1], sum(baseline_cts[1]))
                ), nrow=2
              )
    )$p.value
           )
             )
rownames(pval) = rownames(baseline_cts)
pval
```

```r
# Peak numbers - bar chart

draw_peaks_annot = function(df, c, t, last=F) {
    colnames(df) = c("Region", "Freq")
    df$Freq = as.numeric(df$Freq)
    df$Region = factor(df$Region, levels = c("Intron", "Intergenic", 
        "Promoter", "Exon", "3' UTR", "TTS", "5' UTR"))
    g = ggplot(df, aes(x = Region, y = Freq)) + 
        geom_bar(stat = "identity", fill = c) + 
        theme_bw() + ylab("") + xlab("") + 
        geom_hline(yintercept = 0, size = 0.25) + 
#        ggtitle(t) +
        theme(panel.border = element_blank(), 
              legend.position = "none", 
              panel.grid.minor = element_blank(), 
              panel.grid.major = element_blank(), 
              axis.text.x = element_text(angle = 45, hjust = 1), 
              axis.line = element_line(colour = "black", size = 0.25), 
              text = element_text(size = 6), 
              plot.title = element_text(size = 8))
    if (!last) g = g + theme(axis.text.x = element_blank())
    return(g)
}

df = as.data.frame(cbind(rownames(baseline_cts), baseline_cts[,1]))
p1 = draw_peaks_annot(df, c="black", t="All peaks")

df1 = as.data.frame(cbind(rownames(baseline_cts), baseline_cts[,2]))
df2 = as.data.frame(cbind(rownames(baseline_cts), -baseline_cts[,3]))
p2 = draw_peaks_annot(rbind(df1, df2), c="#F8766D", t="BRQ 48h")

df3 = as.data.frame(cbind(rownames(baseline_cts), baseline_cts[,4]))
df4 = as.data.frame(cbind(rownames(baseline_cts), -baseline_cts[,5]))
p3 = draw_peaks_annot(rbind(df3, df4), c="#00BFC4", t="-E2 48h", last=T)

p1 / p2 / p3
#ggsave("../plots/FigS5_genomic_dist_bars.pdf", width=2, height=2.5)
```

## Calculating promoter signal

### Only need to run once

```r
# get promoters
txdb = TxDb.Mmusculus.UCSC.mm10.knownGene
proms = promoters(genes(txdb), upstream = 2000, downstream = 500)
proms
```

```r
# get overlap between row # of promoter and row # of peak
prom_peak_overlap = as.data.frame(findOverlaps(proms,peak_ranges,ignore.strand=TRUE))
head(prom_peak_overlap)
dim(prom_peak_overlap)
```

```r
# add up norm_expr peak counts for all peaks within each promoter
# need to use log_sum to add log2 read counts correctly
gene_to_peak_sums = list()

log_sum = function(x) {
    if(length(x) == 1) return(x)
    return(log2(sum(sapply(x,function(x) 2**x))))
}

for (gene in unique(prom_peak_overlap$queryHits)) {
    peaks = as.vector(prom_peak_overlap[prom_peak_overlap$queryHits == gene,"subjectHits"])
    gene_to_peak_sums[[gene]] = apply(norm_expr[sapply(peaks, function(x) paste("peak", x, sep='')),],2,log_sum)
}

gene_to_peak_sums = data.frame(matrix(unlist(gene_to_peak_sums), ncol=length(gene_to_peak_sums[[1]]), byrow=TRUE))
rownames(gene_to_peak_sums) = unique(prom_peak_overlap$queryHits)
colnames(gene_to_peak_sums) = colnames(norm_expr)
gene_to_peak_sums = na.omit(gene_to_peak_sums)
print(dim(gene_to_peak_sums))
head(gene_to_peak_sums)
```

```r
# annotate each row with the gene ID 
row_to_gene_id = as.data.frame(proms)[as.integer(rownames(gene_to_peak_sums)),]
head(row_to_gene_id)
```

```r
# re-annotate with gene name
ensembl <- useMart("ensembl", dataset="mmusculus_gene_ensembl")
gene_id_to_name = getBM(attributes=c('entrezgene_id', 'external_gene_name'), 
                      filters = 'entrezgene_id', 
                      values = rownames(row_to_gene_id), 
                      mart = ensembl)
gene_id_to_name = gene_id_to_name[!duplicated(gene_id_to_name$entrezgene_id),]
gene_id_to_name = gene_id_to_name[!duplicated(gene_id_to_name$external_gene_name),]
rownames(gene_id_to_name) = gene_id_to_name$entrezgene_id

rownames(gene_to_peak_sums) = rownames(as.data.frame(proms)[as.integer(rownames(gene_to_peak_sums)),])
gene_to_peak_sums = gene_to_peak_sums[which(rownames(gene_to_peak_sums) %in% gene_id_to_name$entrezgene_id),]
rownames(gene_to_peak_sums) = gene_id_to_name$external_gene_name
head(gene_to_peak_sums)
```

```r
# for each gene, get a df with gene, gene lfc, peak, and peak lfc

get_brq_peaks = function(gene) {
    gene_id = gene_id_to_name[gene_id_to_name$external_gene_name == gene,]$entrezgene_id
    peaks = sapply(gene_peak_overlap[gene_peak_overlap$gene_id == as.character(gene_id),]$subjectHits,
                   function(x) paste("peak", x, sep=''))
    if (length(peaks) == 0) return(data.frame())  # no peaks could be found
    peak_df = norm_expr[peaks[which(peaks %in% rownames(norm_expr))],]
    if (dim(peak_df)[1] == 0) return(data.frame())  # no peaks are within the norm_expr matrix 
    peak_lfc = apply(peak_df[,c("BRQ48_rep1", "BRQ48_rep2")],1,mean) - 
               apply(peak_df[,c("BRQ0_rep1", "BRQ0_rep2")],1,mean)
    peak_avg = apply(peak_df[,c("BRQ0_rep1", "BRQ0_rep2")],1,mean)
                   
    df = cbind(gene, brq48_rna_lfc[gene], rownames(peak_df), peak_lfc, peak_avg)    
    return(df)
}

brq48_enh_peaks = do.call(rbind, lapply(names(brq48_rna_lfc), get_brq_peaks))
brq48_enh_peaks = as.data.frame(brq48_enh_peaks)
colnames(brq48_enh_peaks) = c("gene", "lfc", "peak", "peak_lfc", "peak_avg")
brq48_enh_peaks$lfc = as.numeric(brq48_enh_peaks$lfc)
brq48_enh_peaks$peak_lfc = as.numeric(brq48_enh_peaks$peak_lfc)
```

```r
# for each gene, get a df with gene, gene lfc, peak, and peak lfc
get_e2_peaks = function(gene) {
    gene_id = gene_id_to_name[gene_id_to_name$external_gene_name == gene,]$entrezgene_id
    peaks = sapply(gene_peak_overlap[gene_peak_overlap$gene_id == as.character(gene_id),]$subjectHits,
                   function(x) paste("peak", x, sep=''))
    if (length(peaks) == 0) return(data.frame())
    peak_df = norm_expr[peaks[which(peaks %in% rownames(norm_expr))],]
    if (dim(peak_df)[1] == 0) return(data.frame())
    peak_lfc = apply(peak_df[,c("E2wd48_rep1", "E2wd48_rep2")],1,mean) - 
               apply(peak_df[,c("E2wd0_rep1", "E2wd0_rep2")],1,mean)
    peak_avg = apply(peak_df[,c("E2wd0_rep1", "E2wd0_rep2")],1,mean)
                   
    df = cbind(gene, e2w48_rna_lfc[gene], rownames(peak_df), peak_lfc, peak_avg)    
    return(df)
}

e2w48_enh_peaks = do.call(rbind, lapply(names(e2w48_rna_lfc), get_e2_peaks))
e2w48_enh_peaks = as.data.frame(e2w48_enh_peaks)
colnames(e2w48_enh_peaks) = c("gene", "lfc", "peak", "peak_lfc", "peak_avg")
e2w48_enh_peaks$lfc = as.numeric(e2w48_enh_peaks$lfc)
e2w48_enh_peaks$peak_lfc = as.numeric(e2w48_enh_peaks$peak_lfc)
```

```r
get_deviations = function(data, motif_set) {
  motif_ix <- matchMotifs(motif_set, data,  genome = BSgenome.Mmusculus.UCSC.mm10)
  dev <- computeDeviations(object = data, annotations = motif_ix)
  return(dev)
}

# ordered by variability
get_values = function(data, dev, column_names, num_motifs=-1, pct=FALSE) {
  dev_deviations = deviations(dev)
  dev_zsc = deviationScores(dev)
  
  variability <- computeVariability(dev)
  
  if(num_motifs == -1) num_motifs = dim(variability)[1]
  highly_variable = variability[order(-variability[,"variability"]),][1:num_motifs,]
  dev_highvar = dev_deviations[rownames(highly_variable),]
  colnames(dev_highvar)= column_names
  if(!pct) return(dev_highvar)

  #rownames(dev_highvar) = highly_variable$name
  dev_highvar_pct = t(apply(dev_highvar,1,function(x) (x - min(x))/(max(x) - min(x))))
  return(dev_highvar_pct)
}

# ordered by when the motif is maximallly open
order_by_max = function(dev) {
  dev_order = apply(dev,1,which.max)
  return(dev[order(dev_order),])
}

melt_df = function(data) {
  dev_highvar_melt = melt(data)
  colnames(dev_highvar_melt) = c("Motif", "Timepoint", "Pct")
  return(dev_highvar_melt)
}

chname = function(x) {
  x = sub("XP_", "XP", x)
  x = sub("NP_", "NP", x)
  return(x)
}
```

```r
# brq mouse
cols = c("DMSO24_1", "DMSO24_2", "BRQ8_1", "BRQ8_2", "BRQ16_1", "BRQ16_2", 
         "BRQ24_1", "BRQ24_2", "E2wd24_1", "E2wd24_2")
dev_mouse = get_deviations(fragment_counts, mouse_pwms_v2)
dev_mouse_vals = get_values(fragment_counts, dev_mouse, cols, pct=FALSE)
dev_mouse_pct = get_values(fragment_counts, dev_mouse, cols, pct=TRUE)
rownames(dev_mouse_pct) = sapply(rownames(dev_mouse_pct), 
                                     function(x) {strsplit(chname(x), '_')[[1]][3]})
rownames(dev_mouse_vals) = sapply(rownames(dev_mouse_vals), 
                                      function(x) {strsplit(chname(x), '_')[[1]][3]})
dev_mouse_pct = dev_mouse_pct[!duplicated(rownames(dev_mouse_pct)),]
dev_mouse_vals = dev_mouse_vals[!duplicated(rownames(dev_mouse_vals)),]

write.table(dev_mouse_vals, "../output/210929_bd266_chromvar_mouse.csv", sep=',', quote=FALSE)
```
