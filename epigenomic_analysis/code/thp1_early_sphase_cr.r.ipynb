```r
atac_df = read.table("../data/220802_BD345_THP1_K27ac_PU1_at_atac.txt", header = TRUE, row.names=1)
atac_peak_info = atac_df[,1:5]

atac_df = atac_df[,6:17]
colnames(atac_df) = c("kas1", "kas2", "kau1", "kau2", "kue1", "kue2", "kul1", "kul2", "khe1", "khe2", "khl1", "khl2")

                 # "pas1", "pas2", "pau1", "pau2", "pue1", "pue2", "pul1", "pul2", "phe1", "phe2", "phl1", "phl2",
                 # "igg_hu1", "igg_hu2", "igg_unt1", "igg_unt2")
head(atac_df)
dim(atac_df)
```

```r
# differential peaks
# only considering 5 through 12 because the first four are from a different cell type.
# differential peaks
d0 <- DGEList(subset(atac_df, atac_peak_info$Length < 5000))
d0 <- calcNormFactors(d0, method="TMM")
# cutoff <- 1
# drop <- which(apply(cpm(d0), 1, max) < cutoff)
# d <- d0[-drop,] 
d = d0
dim(d) # number of peaks left
```

```r
ice = c(rep("ice", 2), rep("noice", 2), rep("ice", 4), rep("noice", 4)) 
ice = factor(ice, levels=c("noice", "ice"))

phase = c(rep("async", 4), rep("early", 2), rep("late", 2), rep("early", 2), rep("late", 2))
phase = factor(phase, levels=c("early", "late", "async"))

tx = c(rep("unt", 8), rep("hu", 4))
tx = factor(tx, levels=c("unt", "hu"))

conds = c(rep("other", 4), "ue", "ue", "ul", "ul", "he", "he", "hl", "hl")
conds = factor(conds, levels=c("other", "he", "hl", "ue", "ul"))

# no interactions
mm <- model.matrix(~ ice + conds)

# # with interactions
# mm = model.matrix(~ice + tx*phase)

colnames(mm) = make.names(colnames(mm))

y <- voom(d, mm, plot = F)
fit <- lmFit(y, mm)
head(coef(fit))

# get differentially accessible peaks for each condition
process_res = function(res) {
    tmp <- contrasts.fit(fit, res)
    tmp <- eBayes(tmp)
    res <- topTable(tmp, sort.by = "P", n = Inf)
    return(res)
}
```

```r
# if using conds
contrast.matrix = makeContrasts(HvsU=(condshl+condshe)/2-(condsul+condsue)/2, 
                                ULvsUE=(condsul-condsue),
                                HLvsHE=(condshl-condshe),
                                HLvsUL=condshl-condsul,
                                HEvsUE=condshe-condsue,
                                levels=colnames(coef(fit)))

fit2 = contrasts.fit(fit, contrast.matrix)
efit2 = eBayes(fit2)
results = decideTests(efit2)
#vennDiagram(results)
```

```r
heue = topTable(efit2, coef="HEvsUE", n = Inf)
he_up = subset(heue, logFC > 1 & adj.P.Val < 0.05)  # 3523
he_down = subset(heue, logFC < -1 & adj.P.Val < 0.05)  # 7104

# hlul = topTable(efit2, coef="HLvsUL", n = Inf)
# hl_up = subset(hlul, logFC > 1 & adj.P.Val < 0.05)  # 4166
# hl_down = subset(hlul, logFC < -1 & adj.P.Val < 0.05)  # 2109
```
