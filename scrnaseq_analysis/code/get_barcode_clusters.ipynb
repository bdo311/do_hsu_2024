```python
import numpy as np
import pickle
import pandas as pd
```

```python
bc_to_cells = pickle.load(open("../../files/200630Van/D20-2488-4537L/cells.pkl", 'rb'))
```

```python
bc_to_cells
```

```python
cell_to_bc = {}
for bc in bc_to_cells:
    cells = [x[:16] for x in bc_to_cells[bc]]
    for c in cells:
        cell_to_bc[c + '-1'] = bc[:8]
    
```

```python
cb = pd.DataFrame.from_dict(cell_to_bc, orient='index')
cb
```

```python
cb.to_csv("200814_cell_to_barcode.csv")
```
