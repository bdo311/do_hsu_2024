## Read in raw data

```r
data <- list()
sheetlist <- excel_sheets(path="../data/230313_job2716_Long.XLS")

sample_names = c()
for (metab in sheetlist[1:(length(sheetlist)-2)]){
    tempdf <- read_excel(path="../data/230313_job2716_Long.XLS", sheet = metab, skip=4)
    #print(tempdf["Area"])
    data[[metab]] <- tempdf %>% pull("Area")
    sample_names = tempdf %>% pull("Sample ID")
}

data = as.data.frame(data)
colnames(data) = sheetlist[1:(length(sheetlist)-2)]
data[data == "NF"] = NA
data[] <- lapply(data, function(x) as.numeric(x))

thp1_data = data[1:39,]
sample_names = gsub("BD395-4_sample", "", sample_names[1:39])
print(dim(thp1_data)) 
```

```r
# removing metabolites with >20 NA values, and changing the remaining NA values to 0
thp_data_zero = thp1_data[,apply(thp1_data,2,function(x) sum(is.na(x))) < 20]
thp_data_zero = t(thp_data_zero)
colnames(thp_data_zero) = sample_names
# second_to_last = dim(thp1_data_zero)[1]-1
# thp1_data_zero = rbind(thp1_data_zero[second_to_last:(second_to_last+1),], thp1_data_zero[1:(second_to_last-1),])
                        
#colnames(data_zero) = make.names(data_zero[1,], unique = TRUE)
dim(thp_data_zero)
head(thp_data_zero)
```

```r
thp_data_zero[c("dATP", "dCTP", "dTTP", "ATP", "CTP", "GTP", "UTP"),]
thp_data_zero[c("CMP", "CDP", "CTP"),]
```

```r
std_rows = thp_data_zero[grepl('15N', rownames(thp_data_zero)),]
std_rows
```

```r
# normalize for extraction efficiency -- 15N 
# std_rows = data_zero_mat[grepl('valine-13C5-15N', rownames(data_zero_mat)),]
# std_norm_vals = std_rows

std_rows = thp_data_zero[grepl('15N', rownames(thp_data_zero)),]
std_rows = std_rows[!(rownames(std_rows) %in% c("glycine_13C2-15N", "alanine_13C3-15N", "serine_13C3-15N", "glutamate_13C5-15N")),]  # these have low TICs
std_norm_vals = std_rows/apply(std_rows[,c(1:3)],1,median, na.rm=T)
std_norm_vals = apply(std_norm_vals,2,median, na.rm=T)

thp_data_zero_norm = t(t(thp_data_zero) / std_norm_vals)
# options(repr.plot.width=15,repr.plot.height=5)
# boxplot(std_rows/std_rows[,1])
# abline(h=1)

# std_rows = data_zero_mat_norm[grepl('15N', rownames(data_zero_mat)),]
# boxplot(std_rows/std_rows[,1])
# abline(h=1)

```

```r
# # median normalize
medians = apply(thp_data_zero_norm, 2, function(x) median(x, na.rm=T)) 
thp_data_zero_norm2 = sweep(thp_data_zero_norm, 2, medians, `/`)

# # OR normalize to cell number
# cellnum_df = read_excel("230214_BD395-4_doublings.xlsx", sheet=1, col_names=F)
# cellnums = as.numeric(t(as.matrix(cellnum_df[,2:4])))
# cellnums

# er_data_zero_norm2 = sweep(er_data_zero_norm, 2, cellnums, `/`)
```

```r
# log fold change compared to NTCs 

thp_data_zero_norm2[thp_data_zero_norm2 == 0] = NA
thp_data_zero_lfc = log2(thp_data_zero_norm2/apply(thp_data_zero_norm2[,c(1:3)],1,median, na.rm=T))

thp_data_zero_lfc[is.infinite(thp_data_zero_lfc)] = NA
thp_data_zero_lfc = thp_data_zero_lfc[order(rownames(thp_data_zero_lfc)),]
thp_data_zero_lfc = thp_data_zero_lfc[apply(thp_data_zero_lfc,1,function(x) sum(is.na(x))) < 20,]

var_vals = apply(thp_data_zero_lfc,1,function(x) var(x, na.rm=T))

                 
```

```r
options(repr.plot.width=12,repr.plot.height=18)
col_fun = colorRamp2(c(-6, 0, 6), c("blue", "white", "red"))

thp_data_zero_lfc_highvar = thp_data_zero_lfc[var_vals>0.5,]

#pdf("./220119_bd295_erhoxa9_highvar_heatmap.pdf", width=12, height=18)

d = Heatmap(thp_data_zero_lfc_highvar, cluster_columns = TRUE,
        cluster_rows=TRUE,
        #row_km = 4, row_km_repeats=100,
        column_names_side = "top", col=col_fun,
        na_col="darkgray")

d
```

## Plot nucleotides (THP-1)

```r
thp_sample_to_tx = as.data.frame(cbind(colnames(thp_data_zero_norm2), sapply(colnames(thp_data_zero_norm2), 
                                                             function(x) unlist(strsplit(x, 
                                                                                         split='_', fixed=T))[2])))
colnames(thp_sample_to_tx) = c("sample", "tx")
thp_sample_to_tx[37:39,"tx"] = "rot_hxu"
head(thp_sample_to_tx)
```

```r
plot_vals_thp = function(metabs, xax, conds, w=8, h=14, s=1) {
    df = melt(thp_data_zero_lfc[metabs,])
    df = merge(df, thp_sample_to_tx, by.x = "Var2", by.y="sample")
    df = subset(df, tx %in% conds)
    df$tx = factor(df$tx, levels=conds)
    #print(head(df))

    options(repr.plot.width=w, repr.plot.height=h)
    p = ggplot(data=df, aes(x= !!sym(xax), y=value)) + geom_boxplot() + geom_point(size=s) +  theme_bw() +
                theme(axis.text.x = element_text(angle = 45, hjust=1)) + 
                ylab("log2FC") + geom_hline(yintercept=0, color="grey")
    return(p)
}
```

```r
# dntps = c("dAMP", "dADP", "dATP", "dCDP", "dCTP", "dTMP", "dTDP", "dTTP")
dntps = c("dATP", "dCTP", "dTTP")
plot_vals_thp(dntps, xax = "Var1",
          unique(thp_sample_to_tx$tx), w=16, h=5, s=0.3) + 
          facet_wrap(facets = vars(tx), nrow=1)

#           facet_wrap(facets = vars(tx), nrow=3, scales = "free")

rntps = c("ATP", "CTP", "GTP", "UTP")
plot_vals_thp(rntps, xax = "Var1",
          unique(thp_sample_to_tx$tx), w=16, h=5, s=0.3) + 
          facet_wrap(facets = vars(tx), nrow=1)
```
