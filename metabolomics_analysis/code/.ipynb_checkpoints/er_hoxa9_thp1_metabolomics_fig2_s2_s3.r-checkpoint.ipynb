## THP-1

```r
# removing metabolites with >20 NA values, and changing the remaining NA values to 0
thp1_data_zero = thp1_data[,apply(thp1_data,2,function(x) sum(is.na(x))) < 10]
thp1_data_zero = t(thp1_data_zero)
# second_to_last = dim(thp1_data_zero)[1]-1
# thp1_data_zero = rbind(thp1_data_zero[second_to_last:(second_to_last+1),], thp1_data_zero[1:(second_to_last-1),])
                        
#colnames(data_zero) = make.names(data_zero[1,], unique = TRUE)

thp1_data_zero = thp1_data_zero[,c(1,3:9,11:12)]  # removing problematic ones
dim(thp1_data_zero)
head(thp1_data_zero)
```

```r
# normalize for extraction efficiency -- 15N 
# std_rows = data_zero_mat[grepl('valine-13C5-15N', rownames(data_zero_mat)),]
# std_norm_vals = std_rows

std_rows = thp1_data_zero[grepl('15N', rownames(thp1_data_zero)),]
#std_rows = std_rows[!(rownames(std_rows) %in% c("methionine_13C6-15N")),]  # this was NA in some of the samples
std_norm_vals = std_rows/apply(std_rows[,1:3],1,median, na.rm=T)
std_norm_vals = apply(std_norm_vals,2,median, na.rm=T)

thp1_data_zero_norm = sweep(thp1_data_zero, 2, std_norm_vals, `/`)

# median norm
medians = apply(thp1_data_zero_norm, 2, function(x) median(x, na.rm=T))
thp1_data_zero_norm2 = sweep(thp1_data_zero_norm, 2, medians, `/`)

# options(repr.plot.width=15,repr.plot.height=5)
# boxplot(std_rows/std_rows[,1])
# abline(h=1)

# std_rows = data_zero_mat_norm[grepl('15N', rownames(data_zero_mat)),]
# boxplot(std_rows/std_rows[,1])
# abline(h=1)

```

```r
# log fold change compared to NTCs 

thp1_data_zero_norm2[thp1_data_zero_norm2 == 0] = NA
thp1_data_zero_norm2 = thp1_data_zero_norm2

thp1_data_zero_lfc = log2(thp1_data_zero_norm2/apply(thp1_data_zero_norm2[,1:3],1,median, na.rm=T))
#data_zero_lfc = log2(data_zero_mat_norm2/data_zero_mat_norm2[,2])  # normalize by #2

thp1_data_zero_lfc[is.infinite(thp1_data_zero_lfc)] = NA
#data_zero_lfc[is.nan(data_zero_lfc)] = NA
thp1_data_zero_lfc = thp1_data_zero_lfc[order(rownames(thp1_data_zero_lfc)),]
thp1_data_zero_lfc = thp1_data_zero_lfc[apply(thp1_data_zero_lfc,1,function(x) sum(is.na(x))) < 12,]

var_vals = apply(thp1_data_zero_lfc,1,function(x) var(x, na.rm=T))

                 
```

## ER-Hoxa9

```r
# removing metabolites with >20 NA values, and changing the remaining NA values to 0
er_data_zero = erhoxa9_data[,apply(erhoxa9_data,2,function(x) sum(is.na(x))) < 20]
er_data_zero = t(er_data_zero)
# second_to_last = dim(thp1_data_zero)[1]-1
# thp1_data_zero = rbind(thp1_data_zero[second_to_last:(second_to_last+1),], thp1_data_zero[1:(second_to_last-1),])
                        
#colnames(data_zero) = make.names(data_zero[1,], unique = TRUE)
dim(er_data_zero)
head(er_data_zero)
```

```r
# normalize for extraction efficiency -- 15N 
# std_rows = data_zero_mat[grepl('valine-13C5-15N', rownames(data_zero_mat)),]
# std_norm_vals = std_rows

std_rows = er_data_zero[grepl('15N', rownames(er_data_zero)),]
std_rows = std_rows[!(rownames(std_rows) %in% c("lysine_13C6-15N2", "methionine_13C6-15N", "glycine_13C2-15N")),]  # this was NA in some of the samples

# divide each value by the median of the vehicle values for that metabolite
std_norm_vals = sweep(std_rows, 1, apply(std_rows[,2:4],1,median, na.rm=T), `/`)  

# get the median of all the fold changes across all metabolites for a sample
std_norm_vals = apply(std_norm_vals,2,median, na.rm=T)

# divide all metabolites by that normalization factor
er_data_zero_norm = sweep(er_data_zero, 2, std_norm_vals, `/`)

# median norm for each sample
medians = apply(er_data_zero_norm, 2, function(x) median(x, na.rm=T))
er_data_zero_norm2 = sweep(er_data_zero_norm, 2, medians, `/`)


# options(repr.plot.width=15,repr.plot.height=5)
# boxplot(std_rows/std_rows[,1])
# abline(h=1)

# std_rows = data_zero_mat_norm[grepl('15N', rownames(data_zero_mat)),]
# boxplot(std_rows/std_rows[,1])
# abline(h=1)

```

```r
# log fold change compared to NTCs 

er_data_zero_norm2[er_data_zero_norm2 == 0] = NA
er_data_zero_lfc = log2(er_data_zero_norm2/apply(er_data_zero_norm2[,1:4],1,median, na.rm=T))

to_remove = rownames(er_data_zero_lfc)[grepl("13C", rownames(er_data_zero_lfc))]
to_remove = c(to_remove, "itaconic_acid")
er_data_zero_lfc = subset(er_data_zero_lfc, !(rownames(er_data_zero_lfc) %in% to_remove))

#data_zero_lfc = log2(data_zero_mat_norm2/data_zero_mat_norm2[,2])  # normalize by #2

er_data_zero_lfc[is.infinite(er_data_zero_lfc)] = NA
#data_zero_lfc[is.nan(data_zero_lfc)] = NA
er_data_zero_lfc = er_data_zero_lfc[order(rownames(er_data_zero_lfc)),]
er_data_zero_lfc = er_data_zero_lfc[apply(er_data_zero_lfc,1,function(x) sum(is.na(x))) < 20,]

var_vals = apply(er_data_zero_lfc,1,function(x) var(x, na.rm=T))

                 
```

```r
options(repr.plot.width=12,repr.plot.height=18)
col_fun = colorRamp2(c(-6, 0, 6), c("blue", "white", "red"))

er_data_zero_lfc_highvar = er_data_zero_lfc[var_vals>0.5,]


d = Heatmap(er_data_zero_lfc_highvar, cluster_columns = FALSE,
        cluster_rows=TRUE,
        #row_km = 4, row_km_repeats=100,
        column_names_side = "top", col=col_fun,
        na_col="darkgray")
d

# pdf("./220119_bd295_erhoxa9_highvar_heatmap.pdf", width=12, height=18)
# d
# dev.off()
```

```r
get_lfc_df = function(df) {
    print(head(df,1))
    df = as.data.frame(df[complete.cases(df),])
    diff = apply(df, 1, function(x) median(x[5:8])-median(x[1:4]))
    pvals = apply(df, 1, function(x) t.test(x[5:8], x[1:4])$p.value)
    df$diff = diff
    df$pval = pvals
    df$adj.p.val = -log10(p.adjust(pvals, method="BH"))
                  
    return(df)
}
```

```r
options(repr.plot.width=6, repr.plot.height=6)
hu_vs_veh = get_lfc_df(er_data_zero_lfc[,c(1:4, 9:12)])
print(dim(subset(hu_vs_veh, adj.p.val > 1 & diff > 0.5)))
print(dim(subset(hu_vs_veh, adj.p.val > 1 & diff < -0.5)))

ggplot(hu_vs_veh, aes(x=diff, y=adj.p.val, label=rownames(hu_vs_veh))) + 
geom_vline(xintercept=0, color="gray") + geom_vline(xintercept=c(-0.5,0.5), color="gray", linetype="dashed") + 
geom_hline(yintercept=1, color="gray", linetype="dashed") + 
geom_point() + geom_text_repel(data=subset(hu_vs_veh, adj.p.val > 1), aes(label=rownames(subset(hu_vs_veh, adj.p.val > 1))), force = 120) + 
coord_cartesian(xlim=c(-3,3),ylim=c(0,3)) + theme_bw() +
ggtitle("HU") + xlab("log2 fold change") + ylab("-log10 adjusted p-value") +
theme(panel.grid.minor=element_blank(), panel.grid.major=element_blank())

ggsave("../output/230917_bd295_er_hu.pdf", width=4, height=4)
```

```r
options(repr.plot.width=6, repr.plot.height=6)
aph_vs_veh = get_lfc_df(er_data_zero_lfc[,c(1:4, 13:16)])
print(dim(subset(aph_vs_veh, adj.p.val > 1 & diff > 0.5)))
print(dim(subset(aph_vs_veh, adj.p.val > 1 & diff < -0.5)))

ggplot(aph_vs_veh, aes(x=diff, y=adj.p.val, label=rownames(aph_vs_veh))) + 
geom_vline(xintercept=0, color="gray") + geom_vline(xintercept=c(-0.5,0.5), color="gray", linetype="dashed") + 
geom_hline(yintercept=1, color="gray", linetype="dashed") + 
geom_point() + geom_text_repel(data=subset(aph_vs_veh, adj.p.val > 1), aes(label=rownames(subset(aph_vs_veh, adj.p.val > 1))), force = 120) + 
coord_cartesian(xlim=c(-3,3),ylim=c(0,3)) + theme_bw() +
ggtitle("APH") + xlab("log2 fold change") + ylab("-log10 adjusted p-value") +
theme(panel.grid.minor=element_blank(), panel.grid.major=element_blank())

ggsave("../output/230917_bd295_er_aph.pdf", width=4, height=4)
```
