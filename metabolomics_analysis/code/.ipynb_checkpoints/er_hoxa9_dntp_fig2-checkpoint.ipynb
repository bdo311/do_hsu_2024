## Read in raw data

### ER-Hoxa9

```r
data <- list()
sheetlist <- excel_sheets(path="../data/230322_BD395-6_Long.XLS")

sample_names = c()
for (metab in sheetlist[1:(length(sheetlist)-2)]){
    tempdf <- read_excel(path="../data/230322_BD395-6_Long.XLS", sheet = metab, skip=4)
    #print(tempdf["Area"])
    data[[metab]] <- tempdf %>% pull("Area")
    sample_names = tempdf %>% pull("Sample ID")
}

data = as.data.frame(data)
colnames(data) = sheetlist[1:(length(sheetlist)-2)]
data[data == "NF"] = NA
data[] <- lapply(data, function(x) as.numeric(x))

erhoxa9_data = data[1:21,]
sample_names = gsub("BD395-6_sample", "", sample_names[1:21])
print(dim(erhoxa9_data)) 
```

```r
# removing metabolites with >20 NA values, and changing the remaining NA values to 0
er_data_zero = erhoxa9_data[,apply(erhoxa9_data,2,function(x) sum(is.na(x))) < 5]
er_data_zero = t(er_data_zero)
colnames(er_data_zero) = sample_names
# second_to_last = dim(thp1_data_zero)[1]-1
# thp1_data_zero = rbind(thp1_data_zero[second_to_last:(second_to_last+1),], thp1_data_zero[1:(second_to_last-1),])
                        
#colnames(data_zero) = make.names(data_zero[1,], unique = TRUE)
dim(er_data_zero)
head(er_data_zero)
```

```r
er_data_zero[c("AMP", "ADP", "ATP"),]
er_data_zero[c("CMP", "CDP", "CTP"),]
er_data_zero[c("GMP", "GDP", "GTP"),]
er_data_zero[c("UMP", "UDP", "UTP"),]
```

```r
std_rows = er_data_zero[grepl('15N', rownames(er_data_zero)),]
std_rows
```

```r
# normalize for extraction efficiency -- 15N 
# std_rows = data_zero_mat[grepl('valine-13C5-15N', rownames(data_zero_mat)),]
# std_norm_vals = std_rows

std_rows = er_data_zero[grepl('15N', rownames(er_data_zero)),]
std_rows = std_rows[!(rownames(std_rows) %in% c("glycine_13C2-15N")),]  # these have low TICs
std_norm_vals = std_rows/apply(std_rows[,c(1:3)],1,median, na.rm=T)
std_norm_vals = apply(std_norm_vals,2,median, na.rm=T)

er_data_zero_norm = t(t(er_data_zero) / std_norm_vals)
er_data_zero_norm2 = er_data_zero_norm
# options(repr.plot.width=15,repr.plot.height=5)
# boxplot(std_rows/std_rows[,1])
# abline(h=1)

# std_rows = data_zero_mat_norm[grepl('15N', rownames(data_zero_mat)),]
# boxplot(std_rows/std_rows[,1])
# abline(h=1)

```

```r
# # median normalize
medians = apply(er_data_zero_norm, 2, function(x) median(x, na.rm=T)) 
er_data_zero_norm2 = sweep(er_data_zero_norm, 2, medians, `/`)

# # OR normalize to cell number
# cellnum_df = read_excel("230214_BD395-3_doublings.xlsx", sheet=1, col_names=F)
# cellnums = as.numeric(t(as.matrix(cellnum_df[,2:4])))
# cellnums

# er_data_zero_norm2 = sweep(er_data_zero_norm, 2, cellnums, `/`)

# # OR nothing
# er_data_zero_norm2 = er_data_zero_norm
```

```r
# log fold change compared to NTCs 

er_data_zero_norm2[er_data_zero_norm2 == 0] = NA
er_data_zero_lfc = log2(er_data_zero_norm2/apply(er_data_zero_norm2[,c(1:2)],1,median, na.rm=T))

# normalize by the batch in which the sample was run on the LCMS
er_data_zero_lfc = log2(er_data_zero_norm2/apply(er_data_zero_norm2[,c(1:2)],1,median, na.rm=T))
er_data_zero_lfc[,c(1:7)*3-2] = log2(er_data_zero_norm2[,c(1:7)*3-2]/er_data_zero_norm2[,1])
er_data_zero_lfc[,c(1:7)*3-1] = log2(er_data_zero_norm2[,c(1:7)*3-1]/er_data_zero_norm2[,2])
er_data_zero_lfc[,c(1:7)*3] = log2(er_data_zero_norm2[,c(1:7)*3]/er_data_zero_norm2[,3])


er_data_zero_lfc[is.infinite(er_data_zero_lfc)] = NA
#data_zero_lfc[is.nan(data_zero_lfc)] = NA
er_data_zero_lfc = er_data_zero_lfc[order(rownames(er_data_zero_lfc)),]
er_data_zero_lfc = er_data_zero_lfc[apply(er_data_zero_lfc,1,function(x) sum(is.na(x))) < 20,]

var_vals = apply(er_data_zero_lfc,1,function(x) var(x, na.rm=T))

                 
```

```r
options(repr.plot.width=12,repr.plot.height=9)
col_fun = colorRamp2(c(-6, 0, 6), c("blue", "white", "red"))

er_data_zero_lfc_highvar = er_data_zero_lfc[var_vals>0.5,]

#pdf("./220119_bd295_erhoxa9_highvar_heatmap.pdf", width=12, height=18)

d = Heatmap(er_data_zero_lfc_highvar, cluster_columns = TRUE,
        cluster_rows=TRUE,
        #row_km = 4, row_km_repeats=100,
        column_names_side = "top", col=col_fun,
        na_col="darkgray")

d
```

## Plot nucleotides (ER)

```r
er_sample_to_tx = as.data.frame(cbind(colnames(er_data_zero_norm2), sapply(colnames(er_data_zero_norm2), 
                                                             function(x) unlist(strsplit(x, 
                                                                                         split='_', fixed=T))[2])))
colnames(er_sample_to_tx) = c("sample", "tx")
head(er_sample_to_tx)
```

```r
plot_vals = function(metabs, xax, conds, w=8, h=14, s=1) {
    df = melt(er_data_zero_lfc[metabs,])
    df = merge(df, er_sample_to_tx, by.x = "Var2", by.y="sample")
    df = subset(df, tx %in% conds)
    df$tx = factor(df$tx, levels=conds)
    #print(head(df))

    options(repr.plot.width=w, repr.plot.height=h)
    p = ggplot(data=df, aes(x= !!sym(xax), y=value)) + geom_boxplot() + geom_point(size=s) +  theme_bw() +
                theme(axis.text.x = element_text(angle = 45, hjust=1)) + 
                ylab("log2FC") + geom_hline(yintercept=0, color="grey")
    return(p)
}
```

```r
# dntps = c("dAMP", "dADP", "dATP", "dCDP", "dCTP", "dTMP", "dTDP", "dTTP")
dntps = c("dATP", "dCTP", "dTTP")
plot_vals(dntps, xax = "Var1",
          unique(er_sample_to_tx$tx), w=16, h=5, s=0.3) + 
          facet_wrap(facets = vars(tx), nrow=1)

#           facet_wrap(facets = vars(tx), nrow=3, scales = "free")

rntps = c("ATP", "CTP", "GTP", "UTP")
plot_vals(rntps, xax = "Var1",
          unique(er_sample_to_tx$tx), w=16, h=5, s=0.3) + 
          facet_wrap(facets = vars(tx), nrow=1)
```
