```r
# brq vs dmso

get_lfc_df = function(df) {
    print(head(df,1))
    df = as.data.frame(df[complete.cases(df),])
    diff = apply(df, 1, function(x) median(x[4:6])-median(x[1:3]))
    pvals = apply(df, 1, function(x) t.test(x[4:6], x[1:3])$p.value)
    df$diff = diff
    df$pvals = pvals
    df$adj.p.val = -log10(p.adjust(pvals, method="BH"))
                  
    return(df)
}
                  
```

```r
k562_brq_df = get_lfc_df(k562_data_zero_lfc[,1:6])

print(dim(subset(k562_brq_df, adj.p.val > 1 & diff > 0.5)))
print(dim(subset(k562_brq_df, adj.p.val > 1 & diff < -0.5)))


options(repr.plot.width=6, repr.plot.height=6)


ggplot(k562_brq_df, aes(x=diff, y=adj.p.val, label=rownames(k562_brq_df))) + 
geom_vline(xintercept=0, color="gray") + geom_vline(xintercept=c(-0.5,0.5), color="gray", linetype="dashed") + 
geom_hline(yintercept=1, color="gray", linetype="dashed") + 
geom_point(color="gray") + geom_text_repel(data=subset(k562_brq_df, adj.p.val > 1 & abs(diff) > 1), 
                                           aes(label=rownames(subset(k562_brq_df, adj.p.val > 1 & abs(diff) > 1))), 
                                           force = 60, max.overlap=0) + 
coord_cartesian(xlim=c(-3,3),ylim=c(0,5)) + theme_bw() +
ggtitle("K562 BRQ vs DMSO") + xlab("log2 fold change") + ylab("-log10 adjusted p-value") +
theme(panel.grid.minor=element_blank(), panel.grid.major=element_blank())

ggsave("230919_bd388_k562_brq.pdf", width=4, height=4)
```

```r
k562_hu_df = get_lfc_df(k562_data_zero_lfc[,c(1:3,7:9)])

print(dim(subset(k562_hu_df, adj.p.val > 1 & diff > 0.5)))
print(dim(subset(k562_hu_df, adj.p.val > 1 & diff < -0.5)))


options(repr.plot.width=6, repr.plot.height=6)


ggplot(k562_hu_df, aes(x=diff, y=adj.p.val, label=rownames(k562_hu_df))) + 
geom_vline(xintercept=0, color="gray") + geom_vline(xintercept=c(-0.5,0.5), color="gray", linetype="dashed") + 
geom_hline(yintercept=1, color="gray", linetype="dashed") + 
geom_point(color="gray") + geom_text_repel(data=subset(k562_hu_df, adj.p.val > 1 & abs(diff) > .5), 
                                           aes(label=rownames(subset(k562_hu_df, adj.p.val > 1 & abs(diff) > .5))), 
                                           force = 60, max.overlap=0) + coord_cartesian(xlim=c(-3,3),ylim=c(0,5)) + theme_bw() +
ggtitle("K562 HU vs DMSO") + xlab("log2 fold change") + ylab("-log10 adjusted p-value") +
theme(panel.grid.minor=element_blank(), panel.grid.major=element_blank())

ggsave("230919_bd388_k562_hu.pdf", width=4, height=4)
```

```r
plot_vals = function(metabs, xax, conds, w=8, h=14, s=1) {
    df = melt(as.matrix(thp1_data_zero_lfc[metabs,]))
    df = merge(df, thp1_sample_to_tx, by.x = "Var2", by.y="sample")
    df = subset(df, tx %in% conds)
    df$tx = factor(df$tx, levels=conds)
    #print(head(df))

    options(repr.plot.width=w, repr.plot.height=h)
    p = ggplot(data=df, aes(x= !!sym(xax), y=value)) + geom_boxplot() + geom_point(size=s) +  theme_bw() +
                theme(axis.text.x = element_text(angle = 45, hjust=1)) + 
                ylab("log2FC") + geom_hline(yintercept=0, color="grey")
    return(p)
}
```
