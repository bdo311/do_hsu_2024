### THP-1

```r
# removing metabolites with >10 NA values, and changing the remaining NA values to 0
thp1_data_zero = thp1_data[,apply(thp1_data,2,function(x) sum(is.na(x))) < 10]
thp1_data_zero = t(thp1_data_zero)
# second_to_last = dim(thp1_data_zero)[1]-1
# thp1_data_zero = rbind(thp1_data_zero[second_to_last:(second_to_last+1),], thp1_data_zero[1:(second_to_last-1),])
                        
#colnames(data_zero) = make.names(data_zero[1,], unique = TRUE)
dim(thp1_data_zero)
head(thp1_data_zero)
```

```r
# normalize for extraction efficiency -- 15N 
# std_rows = data_zero_mat[grepl('valine-13C5-15N', rownames(data_zero_mat)),]
# std_norm_vals = std_rows

std_rows = thp1_data_zero[grepl('15N', rownames(thp1_data_zero)),]
std_rows = std_rows[!(rownames(std_rows) %in% c("lysine_13C6-15N2", "methionine_13C6-15N", "glycine_13C2-15N")),]  # this was NA in some of the samples
std_norm_vals = std_rows/apply(std_rows[,1:3],1,median, na.rm=T)
std_norm_vals = apply(std_norm_vals,2,median, na.rm=T)

thp1_data_zero_norm = sweep(thp1_data_zero, 2, std_norm_vals, `/`)

# median norm
medians = apply(thp1_data_zero_norm, 2, function(x) median(x, na.rm=T)) 
thp1_data_zero_norm2 = sweep(thp1_data_zero_norm, 2, medians, `/`)

# options(repr.plot.width=15,repr.plot.height=5)
# boxplot(std_rows/std_rows[,1])
# abline(h=1)

# std_rows = data_zero_mat_norm[grepl('15N', rownames(data_zero_mat)),]
# boxplot(std_rows/std_rows[,1])
# abline(h=1)

```

```r
# log fold change compared to NTCs 

thp1_data_zero_norm2[thp1_data_zero_norm2 == 0] = NA
thp1_data_zero_lfc = log2(thp1_data_zero_norm2/apply(thp1_data_zero_norm2[,1:3],1,median, na.rm=T))
#data_zero_lfc = log2(data_zero_mat_norm2/data_zero_mat_norm2[,2])  # normalize by #2

thp1_data_zero_lfc[is.infinite(thp1_data_zero_lfc)] = NA
to_remove = rownames(thp1_data_zero_lfc)[grepl("13C", rownames(thp1_data_zero_lfc))]
to_remove = c(to_remove, "itaconic_acid")
thp1_data_zero_lfc = subset(thp1_data_zero_lfc, !(rownames(thp1_data_zero_lfc) %in% to_remove))
#data_zero_lfc[is.nan(data_zero_lfc)] = NA
thp1_data_zero_lfc = thp1_data_zero_lfc[order(rownames(thp1_data_zero_lfc)),]
thp1_data_zero_lfc = thp1_data_zero_lfc[apply(thp1_data_zero_lfc,1,function(x) sum(is.na(x))) < 12,]

var_vals = apply(thp1_data_zero_lfc,1,function(x) var(x, na.rm=T))

                 
```

```r
get_lfc_df = function(df) {
    print(head(df,1))
    df = as.data.frame(df[complete.cases(df),])
    diff = apply(df, 1, function(x) median(x[4:6])-median(x[1:3]))
    pvals = apply(df, 1, function(x) t.test(x[4:6], x[1:3])$p.value)
    df$diff = diff
    df$pval = pvals
    df$adj.p.val = -log10(p.adjust(pvals, method="BH"))
                  
    return(df)
}
```

```r
brq_vs_veh = get_lfc_df(thp1_data_zero_lfc[,c(1:3, 4:6)])
print(dim(subset(brq_vs_veh, adj.p.val > 1 & diff > 0.5)))
print(dim(subset(brq_vs_veh, adj.p.val > 1 & diff < -0.5)))


options(repr.plot.width=6, repr.plot.height=6)


ggplot(brq_vs_veh, aes(x=diff, y=adj.p.val, label=rownames(brq_vs_veh))) + 
geom_vline(xintercept=0, color="gray") + geom_vline(xintercept=c(-0.5,0.5), color="gray", linetype="dashed") + 
geom_hline(yintercept=1, color="gray", linetype="dashed") + 
geom_point(color="gray") + geom_text_repel(data=subset(brq_vs_veh, adj.p.val > 1 & abs(diff) > .5), 
                                           aes(label=rownames(subset(brq_vs_veh, adj.p.val > 1 & abs(diff) > .5))), 
                                           force = 60) + coord_cartesian(xlim=c(-3,3),ylim=c(0,5)) + theme_bw() +
ggtitle("THP-1 BRQ vs DMSO") + xlab("log2 fold change") + ylab("-log10 adjusted p-value") +
theme(panel.grid.minor=element_blank(), panel.grid.major=element_blank())
ggsave("../output/230919_bd376_thp1_brq.pdf", width=4, height=4)
```

```r
plot_vals = function(metabs, xax, conds, w=8, h=14, s=1) {
    df = melt(thp1_data_zero_lfc[metabs,])
    df = merge(df, thp1_sample_to_tx, by.x = "Var2", by.y="sample")
    df = subset(df, tx %in% conds)
    df$tx = factor(df$tx, levels=conds)
    #print(head(df))

    options(repr.plot.width=w, repr.plot.height=h)
    p = ggplot(data=df, aes(x= !!sym(xax), y=value)) + geom_boxplot() + geom_point(size=s) +  theme_bw() +
                theme(axis.text.x = element_text(angle = 45, hjust=1)) + 
                ylab("log2FC") + geom_hline(yintercept=0, color="grey")
    return(p)
}
```

```r
#dntps = c("dAMP", "dADP", "dATP", "dCMP", "dCDP", "dCTP", "dTMP", "dTDP", "dTTP")
dntps = c("dATP", "dCTP", "dTTP")
plot_vals(dntps, xax = "Var1",
          unique(thp1_sample_to_tx$tx), w=12, h=4, s=0.3) + 
          facet_wrap(facets = vars(tx), nrow=1)

#           facet_wrap(facets = vars(tx), nrow=3, scales = "free")

# rntps = c("AMP", "ADP", "ATP", "CMP", "CDP", "CTP",
#             "GMP", "GDP", "GTP", "UMP", "UDP", "UTP")
rntps = c("ATP", "CTP", "GTP", "UTP")
plot_vals(rntps, xax = "Var1",
          unique(thp1_sample_to_tx$tx), w=12, h=4, s=0.3) + 
          facet_wrap(facets = vars(tx), nrow=1)
```

```r
er_sample_to_tx = as.data.frame(cbind(colnames(er_data_zero), sapply(colnames(er_data_zero), 
                                                             function(x) unlist(strsplit(x, 
                                                                                         split='__', fixed=T))[2])))
colnames(er_sample_to_tx) = c("sample", "tx")
head(er_sample_to_tx)
```

```r
plot_vals_er = function(metabs, xax, conds, w=8, h=14, s=1) {
    df = melt(er_data_zero_lfc[metabs,])
    df = merge(df, er_sample_to_tx, by.x = "Var2", by.y="sample")
    df = subset(df, tx %in% conds)
    df$tx = factor(df$tx, levels=conds)
    #print(head(df))

    options(repr.plot.width=w, repr.plot.height=h)
    p = ggplot(data=df, aes(x= !!sym(xax), y=value)) + geom_boxplot() + geom_point(size=s) +  theme_bw() +
                theme(axis.text.x = element_text(angle = 45, hjust=1)) + 
                ylab("log2FC") + geom_hline(yintercept=0, color="grey")
    return(p)
}
```

```r
dntps = c("dAMP", "dADP", "dATP", "dCMP", "dCDP", "dCTP", "dTMP", "dTDP", "dTTP")
plot_vals_er(dntps, xax = "Var1",
          unique(er_sample_to_tx$tx), w=16, h=10, s=0.3) + 
          facet_wrap(facets = vars(tx), nrow=2)

#           facet_wrap(facets = vars(tx), nrow=3, scales = "free")

nucleosides = c("adenine", "adenosine", "cytidine", "hypoxanthine", "inosine", "guanine", 
                "guanosine", "uridine", "uracil")
plot_vals_er(nucleosides, xax = "Var1",
          unique(er_sample_to_tx$tx), w=16, h=10, s=0.3) + 
          facet_wrap(facets = vars(tx), nrow=2)
# plot_vals(nucleosides,
#           unique(sample_to_tx$kd), w=6, h=10, s=0.3) + 
#           facet_grid(rows = vars(Var2), scales = "free")

rntps = c("AMP", "ADP", "ATP", "CMP", "CDP", "CTP",
            "GMP", "GDP", "GTP", "UMP", "UDP", "UTP")
plot_vals_er(rntps, xax = "Var1",
          unique(er_sample_to_tx$tx), w=16, h=10, s=0.3) + 
          facet_wrap(facets = vars(tx), nrow=2)
```

```r
er_sample_to_tx_subset = subset(er_sample_to_tx, tx %in% c("VEH", "BRQ", "HU", "LTX", "BRQ_LTX"))
dntps = c("ATP", "CTP", "GTP", "UTP", "dATP", "dCTP", "dTTP")
plot_vals_er(dntps, xax = "tx",
          unique(er_sample_to_tx_subset$tx), w=12, h=3, s=0.3) + 
          facet_wrap(facets = vars(Var1), nrow=1) + ylab("Relative abundance") + xlab("")
ggsave("230606_bd376_brqltx_revision.pdf", width=7, height=2.5)
```
